```python
import json 
import pandas as pd
import requests
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
```

```python
def get_data(url):
    response = requests.get(url)  
    d = response.json()
    # Normalizing data
    total_df = pd.json_normalize(d, errors='ignore')
    return total_df
```

```python
def extract_column(df,column_name):
    df = df.explode(column_name)
    df = pd.json_normalize(df[column_name])
    return df
```

```python
# Define the URL
url='https://data.covid19india.org/v4/min/data.min.json'
# Call the get_data() to retrieve the URL JSON file data
dataset = get_data(url)
```
